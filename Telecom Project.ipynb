```python
print("Mean:",round(G['age'].mean(),2))
print("Std Dev:",round(G['age'].std(),2))

Q1 = G['age'].quantile(0.25)
Q2 = G['age'].quantile(0.5)
Q3 = G['age'].quantile(0.75)
print("Q1:",Q1)
print("Q2:",Q2)
print("Q3:",Q3)

IQR = Q3-Q1
print("IQR:",IQR)

UL = Q3 + 1.5 * IQR
LL = Q1 - 1.5 * IQR
print("UL:",UL)
print("LL:",LL)
```

```python
def get_lat_long(city):                # get lat, long from city
    location = geolocator.geocode(city)
    if location is None:
        return None
    else:
        return location.latitude, location.longitude
```
